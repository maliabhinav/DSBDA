```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/user/OneDrive/Documents/DSBDA/anti dataset.csv")
data
```

## * Equal width partitioning

```python
data_equalPart = data.copy()

bins = [0, 20, 40, 60, 80, 100]

bin_labels = ['Bin 1', 'Bin 2', 'Bin 3', 'Bin 4', 'Bin 5']

data_equalPart['Equal Width Bin'] = pd.cut(data_equalPart['V1'], bins=bins, labels=bin_labels, include_lowest=True)

data_equalPart
```

## *Equal frequency partitioning 

```python
data_equalFreq = data.copy()

sorted_values = data_equalFreq['V1'].sort_values()

num_bins = 40 

# Perform equal frequency partitioning
bin_labels = ['Bin {}'.format(i+1) for i in range(num_bins)]
data_equalFreq['Equal Frequency Bin'] = pd.qcut(sorted_values[:40], q=num_bins, labels=bin_labels)

data_equalFreq
```

## *Dividing data into 2 groups based on T1 virus quantity

```python
group_1 = data[data['V1'] < 50]
group_2 = data[data['V1'] >= 50]

# Display the two groups
print("Group 1 (Lower V1 virus):")
print(group_1)
print("\nGroup 2 (Higher V1 virus):")
print(group_2)
```

## *Finding mean, median and std for each group

```python
group_1.describe().transpose()
#group_1_mean = group_1['V1'].mean()
#group_1_median = group_1['V1'].median()
#group_1_std = group_1['V1'].std()
```

```python
group_2.describe().transpose()
#group_2_mean = group_2['V1'].mean()
#group_2_median = group_2['V1'].median()
#group_2_std = group_2['V1'].std()
```

## *Boxplot of T-virus measure for each group

```python
# Draw a boxplot graph for 'vaccine effectiveness percent' for Group 1
group_1.boxplot(column='V1')
plt.title('Boxplot of T1 Virus for Group 1')
plt.xlabel('Group 1')
plt.ylabel('V1')
plt.show()

group_2.boxplot(column='V1')
plt.title('Boxplot of T1 Virus for Group 2')
plt.xlabel('Group 2')
plt.ylabel('V1')
plt.show()
```

## *Normalization of 'ANTI_A' and 'ANTI_B' Columns for Group_1

```python
# Calculate the minimum and maximum values of 'Antigen A' and 'Antigen B' for Group 1
min_antigen_a = group_1['ANTI_A'].min()
max_antigen_a = group_1['ANTI_A'].max()
min_antigen_b = group_1['ANTI_B'].min()
max_antigen_b = group_1['ANTI_B'].max()

# Apply min-max normalization to 'Antigen A' and 'Antigen B' columns for Group 1
group_1['ANTI_A Normalized'] = (group_1['ANTI_A'] - min_antigen_a) / (max_antigen_a - min_antigen_a)
group_1['ANTI_B Normalized'] = (group_1['ANTI_B'] - min_antigen_b) / (max_antigen_b - min_antigen_b)

# Display the DataFrame with normalized values
print(group_1[['ANTI_A', 'ANTI_B', 'ANTI_A Normalized', 'ANTI_B Normalized']])
```

## * Normalization of 'ANTI_A' and 'ANTI_B' Columns for Group_2

```python
# Calculate the minimum and maximum values of 'Antigen A' and 'Antigen B' for Group 2
min_antigen_a = group_2['ANTI_A'].min()
max_antigen_a = group_2['ANTI_A'].max()
min_antigen_b = group_2['ANTI_B'].min()
max_antigen_b = group_2['ANTI_B'].max()

# Apply min-max normalization to 'Antigen A' and 'Antigen B' columns for Group 1
group_2['ANTI_A Normalized'] = (group_2['ANTI_A'] - min_antigen_a) / (max_antigen_a - min_antigen_a)
group_2['ANTI_B Normalized'] = (group_2['ANTI_B'] - min_antigen_b) / (max_antigen_b - min_antigen_b)

# Display the DataFrame with normalized values
print(group_2[['ANTI_A', 'ANTI_B', 'ANTI_A Normalized', 'ANTI_B Normalized']])
```

## *Scatterplot for 'ANTI_A' and 'ANTI_B' for Each Group

```python
plt.scatter(group_1['ANTI_A Normalized'], group_1['ANTI_B Normalized'])
plt.title('Scatter Plot of Normalized ANTI_A and ANTI_B for Group 1')
plt.xlabel('Normalized ANTI_A')
plt.ylabel('Normalized ANTI_B')
plt.show()

plt.scatter(group_2['ANTI_A Normalized'], group_2['ANTI_B Normalized'])
plt.title('Scatter Plot of Normalized ANTI_A and ANTI_B for Group 2')
plt.xlabel('Normalized ANTI_A')
plt.ylabel('Normalized ANTI_B')
plt.show()
```

## *Pearson Correlation Coefficient between Normalized 'ANTI_A' and 'ANTI_B' for Each  Group

```python
# Calculate the Pearson correlation coefficient between the normalized 'Antigen A' and 'Antigen B'
correlation_coefficient = group_1['ANTI_A Normalized'].corr(group_1['ANTI_B Normalized'])
print("Pearson Correlation Coefficient for Group 1:", correlation_coefficient)

correlation_coefficient = group_2['ANTI_A Normalized'].corr(group_2['ANTI_B Normalized'])
print("Pearson Correlation Coefficient for Group 2:", correlation_coefficient)
```

 For group_1, normalized 'ANTI_A' and 'ANTI_B' values show a weak or negative correlation, it suggests that the medicines have different effects in fighting the virus. For group_2, normalized 'ANTI_A' and 'ANTI_B' values show a strong positive correlation (close to +1), it suggests that both medicines have a similar effect in fighting the virus. In this case, using both medicines together may provide better results.
