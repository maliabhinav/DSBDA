```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/user/OneDrive/Documents/DSBDA/Salary Dataset.csv")
data
```

```python
data.head()
```

```python
data.replace('?', np.NAN, inplace=True)
data.head()
```

```python
missing_values = data.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
data.dropna(inplace=True)
data
```

```python
data.describe().transpose()
```

```python
# Filter out rows where 'Salary' column does not match the expected format
data = data[data['Salary'].str.contains('₹') & data['Salary'].str.contains('/yr')]

# Remove extra characters and symbols from the 'Salary' column and convert it to integer
data['Salary'] = data['Salary'].str.replace('₹', '').str.replace(',', '').str.replace('/yr', '').astype(int)

# Check the first few rows of the DataFrame after conversion
data
```

```python
data.describe().transpose()
```

```python
# Group data based on 'Job Title' and calculate statistics for 'Salary'
stats_by_job_title = data.groupby('Job Title')['Salary'].agg(['mean', 'median', 'min', 'max', 'std'])

# Set the display format for mean values
pd.options.display.float_format = '{:.2f}'.format

# Display the statistics for each category in 'Job Title'
stats_by_job_title
```

The reason for getting 'NaN' (Not a Number) values in the 'std' column is due to the presence of only a single value or very few values in some groups for the 'Salary' variable.

The standard deviation (std) is a measure of the amount of variation or dispersion in a set of values. When calculating the standard deviation for a group that has only one or very few values, it results in 'NaN' because there isn't enough data to compute a meaningful standard deviation.

## Providing summary statistics for grouped data using bar graph

```python
# Sort the DataFrame by mean salary
stats_by_job_title = stats_by_job_title.sort_values(by='mean', ascending=False)

# Create a bar chart with error bars for mean salary and standard deviation
plt.figure(figsize=(10, 6))
plt.bar(stats_by_job_title.index, stats_by_job_title['mean'], yerr=stats_by_job_title['std'], capsize=5)
plt.xticks(rotation=90)
plt.xlabel('Job Title')
plt.ylabel('Salary (LPA)')
plt.title('Mean Salary and Standard Deviation by Job Title')
plt.tight_layout()
plt.show()
```
