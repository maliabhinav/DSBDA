```python
import pandas as pd
```

```python
import numpy as np
```

```python
dataframe = pd.read_csv("EduData.csv")
```

```python
dataframe
```

```python
#dealing with missing value of topic column by replacing NaN with most common value i.e. mode
dataframe = dataframe.replace("?",np.nan)
```

```python
dataframe.isnull().sum()
```

```python
mode = dataframe["Topic"].mode()
dataframe["Topic"] = dataframe["Topic"].fillna(mode[0])
```

```python
dataframe.isnull().sum()
```

```python
#Removing rows of relation and gradeID columns
dataframe = dataframe.dropna(subset = "Relation")
dataframe = dataframe.dropna(subset = "GradeID")
```

```python
dataframe.isnull().sum()
```

```python
#replacing raisehands missing values with mean
dataframe["raisedhands"] = dataframe["raisedhands"].astype(float)
```

```python
mean = dataframe["raisedhands"].mean(skipna = True)
dataframe["raisedhands"] = dataframe["raisedhands"].fillna(mean)
```

```python
dataframe.isnull().sum()
```

```python
dataframe["Semester"].value_counts()
```

```python
dataframe = dataframe.drop(dataframe[dataframe["Semester"] == "T"].index)
```

```python
dataframe["Semester"].value_counts()
```

```python
dataframe["ParentAnsweringSurvey"].value_counts()
```

```python
dataframe = dataframe.drop(columns = ["ParentAnsweringSurvey"])
```

```python
dataframe
```

```python
dataframe["raisedhands"] = dataframe["raisedhands"]/dataframe["raisedhands"].max()
```

```python
dataframe["VisITedResources"] = dataframe["VisITedResources"]/dataframe["VisITedResources"].max()
```

```python
dataframe
```

```python
dataframe["GradeID"].value_counts()
```

```python
list2=list(dataframe["GradeID"].unique())
```

```python
list2.sort()
```

```python
list2
```

```python
i=0
for values in list2:
    dataframe["GradeID"] = dataframe["GradeID"].replace(values,i)
    i+=1
dataframe
```

```python
absentDays = list(dataframe["StudentAbsenceDays"].unique())
```

```python
i=0
for days in absentDays:
    dataframe["StudentAbsenceDays"] = dataframe["StudentAbsenceDays"].replace(days,i)
    i+=1
dataframe
```
